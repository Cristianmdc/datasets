```python
# import yfinance, pandas and os
import yfinance as yf
import pandas as pd
import os
from tqdm import tqdm

a = "CVX"
z = yf.Ticker(a)
esg_data = pd.DataFrame.transpose(z.sustainability)
esg_data['company_ticker'] = str(z.ticker)


tickerlist = ['COP','AR','XOM','PXD','COG',
       'CNQ','CVE','XEC','CLR','CPG','DVN',
       'EOG','OXY','MRO','RRC','APA','EQT',
       'HES','IMO','AAPL','TSLA','JPM','GOOG']

for i in tqdm(tickerlist):
    var = yf.Ticker(i)
    try:
        if var.sustainability is not None:
            df = pd.DataFrame.transpose(var.sustainability)
            df['company_ticker'] = str(var.ticker)
            esg_data = esg_data.append(df)
    except IndexError:
        pass
```

```python
esg_data.columns
```

```python
data = esg_data[['company_ticker','esgPerformance','environmentScore','socialScore','governanceScore','totalEsg']]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

#Top 5 Companies with highest total ESG score
top_5 = data.sort_values(by="totalEsg", ascending=False).head()
top_5.plot(x="company_ticker", y="totalEsg", kind="bar", color='green')
plt.title("Top 5 Companies with highest total ESG score")
plt.show()

#Top 5 Companies with lowest total ESG score
bottom_5 = data.sort_values(by="totalEsg", ascending=False).tail()
bottom_5.plot(x="company_ticker", y="totalEsg", kind="bar", color='red')
plt.title("Top 5 Companies with lowest total ESG score")
plt.show()
```

```python
#Top 5 Companies with highest environmental score
top_5 = data.sort_values(by="environmentScore", ascending=False).head()
top_5.plot(x="company_ticker", y="environmentScore", kind="bar", color='orange')
plt.title("Top 5 Companies with highest environmental score")
plt.show()

#Top 5 Companies with lowest environmental score
bottom_5 = data.sort_values(by="environmentScore", ascending=False).tail()
bottom_5.plot(x="company_ticker", y="environmentScore", kind="bar", color='red')
plt.title("Top 5 Companies with lowest environmental score")
plt.show()
```
