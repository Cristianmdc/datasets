```python
import pandas as pd
import yfinance as yf
from ta import add_all_ta_features
import datetime
```

```python
#pull historical data
def get_stock_data(stock):

    NUM_DAYS = 729    # The number of days of historical data to retrieve
    INTERVAL = '1h'
    symbol = str(stock)   # Symbol of the desired stock

    #define start & dates
    start = (datetime.date.today() - datetime.timedelta( NUM_DAYS ) )
    end = datetime.datetime.today()
    
    #pull data
    df = yf.download(symbol, start=start, end=end, interval=INTERVAL)
    return df
```

```python
df = get_stock_data('JPM')
df
```

```python
#get ta-lib indicators
def get_indicators(df):
    data_ind = add_all_ta_features(df, open="Open", high="High", low="Low", 
                close="Close", volume="Volume")
    return data_ind
```

```python
get_indicators(df)
```
